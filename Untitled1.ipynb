# Other 2-D Graphs

```python
n=np.array([0,1,2,3,4,5])

fig, axes=plt.subplots(1,4, figsize=(16,5))

axes[0].set_title('Scatter Plot')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('Step Plot')
axes[1].step(n,n**2, lw=2)

axes[2].set_title('Bar Plot')
axes[2].bar(n, n**2, align="center", width=0.9, alpha=0.5)

axes[3].set_title('Fill_between Plot')
axes[3].fill_between(x, x**2,x**3, color='green', alpha=0.5)
```

```python
#Draw a radar chart

fig=plt.figure(figsize=(6,6))

axes=fig.add_axes([0.0,0.0,1,1], polar=True)
t =np.linspace(0, 2 * np.pi, 100)
axes.plot(t,t*2, color='blue',lw=3)
```

```python
#Draw Historgram

n=np.random.randn(100000)
fig, axes =plt.subplots(1,2,figsize=(12,4))

axes[0].set_title('Default Histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram')
axes[1].hist(n, cumulative=True, bins=100)
```

```python
#Draw Contour Map


import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x= np.arange(-3.0,3.0, delta)
y= np.arange(-2.0,2.0, delta)
X,Y = np.meshgrid(x,y)
Z1=np.exp(-X**2 -Y**2)
Z2=np.exp(-(X-1)**2- (Y-1)**2)
Z=(Z1-Z2)*2


print(X)
print(Y)
```

```python
fig, ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS, inline=1,  fontsize=10)
ax.set_title('Contour map')
```

```python
#Draw 3D surface image


from mpl_toolkits.mplot3d.axes3d import Axes3D

fig=plt.figure(figsize=(14,6))

#Specify 3D graphics to draw with projections='3d'
ax=fig.add_subplot(1,1,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4,cstride=4, linewidth=0)
```

# PRACTICE EXAMPLE
Write a python programming to create a pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt
#data to plot

languages= 'Python','Java','PHP','Javascript','C#','C++'
popularity= [22.2,17.6,8.8,8,7.7,6.7]
colors=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b']

```

```python
#explode 1st slice
explode=(0.1,0,0,0,0,0)
plt.pie(popularity, explode=explode, labels=languages, colors=colors, 
autopct='%1.1f%%', shadow=True, startangle=90)
```
